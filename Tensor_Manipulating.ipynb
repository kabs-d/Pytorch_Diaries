```python
# Other useful tensor Operations
x = torch.tensor([[-1,2,3], [2,3,4], [3,4,5]])
sum_x = torch.sum(x, dim =1)
print(sum_x)
values, indices = torch.max(x,dim =0)
print(values, indices)
values, indices = torch.min(x, dim = 1)
print(values, indices)
abs_x = torch.abs(x)
print(abs_x)
z = torch.argmax(x, dim =0)
print(z)
z=torch.argmin(x, dim=1)
print(z)
mean_x = torch.mean(x.float(), dim=0)
print(mean_x)
z = torch.eq(x, torch.eye(3))
print(z)
sorted_x, indices = torch.sort(x, dim =0, descending = False)
print(sorted_x)
z= torch.clamp(x, min= 0)
print(z)
x = torch.tensor([1,0,1,1,1], dtype = torch.bool)
z = torch.any(x)
print(z)
z=torch.all(x)
print(z)
```

TENSOR INDEXING

```python
batch_size = 10
features = 25
x = torch.rand((batch_size, features))
print(x[0].shape)
print(x[:,0].shape)
print(x[2, 0:10])
x[0,0] = 100
```

```python
#Fancy Indexing
x = torch.arange(10)
indices = [2,5,8]
print(x[indices])

x = torch.rand((3, 5))
rows = torch.tensor([1,0])
cols = torch.tensor([4,0])
print(x)
print(x[rows, cols])
```

```python
#Advanced Indexing
x = torch.arange(10)
print(x[(x<2) | (x>8)])
print(x[(x<2) & (x>8)])
print(x[x.remainder(2) == 0])


# useful operations
print(torch.where(x>5, x, x*2))
print(torch.tensor([0,0,1,2,2,3,4]).unique())
print(x.numel())
```
